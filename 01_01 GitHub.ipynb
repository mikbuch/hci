# 01.01 GitHub

Instrukacja jak zacząć naszą przygodę z GitHub-em

## Zakładamy konto na GitHub-ie

Jeśli ktoś jeszcze takowego nie posiada, kierujemy się pod [ten](https://github.com/join) link.

## Tworzymy repozytorium

Używamy do tego przeglądarki internetowej na naszym koncie [jak stworzyć repozytorium](https://help.github.com/en/articles/create-a-repo). Zaznacz proszę "initialize with README.md" i napisz dwa zdania o swoim repozytorium.

## Obsługa repozytorium przez dedykowane programy

Ściągamy oprogramowanie Git:
 * [wersję Desktop](https://desktop.github.com/), oraz
 * [Git Bash](https://git-scm.com/downloads) (wersję konsolową).
 
## Podstawowe komendy git-a

Przećwiczmy posługiwanie się konsolą. Najczęściej stosowane komendy git-a to:

### clone

Ściąga repozytorium na nasz komputer:

```bash
git clone https://github.com/mikbuch/hci
```

### status

Wyświetla status naszego repozytorium, m.in.:
 * Które pliki zostały zmodyfikowane
 * Które pliki nie zostały jeszcze dodane do git-a

Komenda:

```bash
git status
```

Przykładowy output:
```bash
➜  hci git:(master) ✗ git status
On branch master
Your branch is up to date with 'origin/master'.

Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)                                   
  (use "git checkout -- <file>..." to discard changes in working directory)                    

        modified:   README.md

Untracked files:
  (use "git add <file>..." to include in what will be committed)                               

        01_GitHub_Jupyter.ipynb

no changes added to commit (use "git add" and/or "git commit -a") 
```

Komenda pokazuje, iż:
 * plik "README.md" został zmodyfikowany
 * plik "01_GitHub_Jupyter.ipynb" nie został jeszcze dodany do repozytorium (do śledzenia zmian).


### add

Aby dodać plik stosujemy komendę `add`:

```bash
git add 01_GitHub_Jupyter.ipynb
```

Aby dodać wszystkie nieśledzone pliki stosujemy następującą komendę:

```bash
git add --all
```

Sprawdzamy status:

```bash
➜  hci git:(master) ✗ git status                                                               
On branch master
Your branch is up to date with 'origin/master'.

Changes to be committed:
  (use "git reset HEAD <file>..." to unstage)

        new file:   01_GitHub_Jupyter.ipynb
```

Plik został dodany, teraz należy zatwierdzić zmiany, które w nim zrobiliśmy. W naszym przypadku są to pierwsze zmiany w tym pliku, ale je też musimy zaakceptować.

### commit

Aby zaaceptować zmiany używamy komendy `commit`:

```bash
git commit 01_GitHub_Jupyter.ipynb -m "Moje pierwsze zmiany do tego pliku"
```

Możemy też, zamiast akceptować zmiany dla każdego pliku osobno, zaakceptować wszystkie zmiany:

```bash
git commit -a -m "Niniejszym zatwierdzam wszystkie zmiany we wszystkich plikach"
```

### push

W końcu możemy "wypchnąć" nasze zmiany na serwer:

```bash
git push
```

## Ćwiczenie

Utwórz własne repozytorium zawierające przykładowy kod, który stworzyłeś/stworzyłaś na zajęciach z programowania.

## Uwagi końcowe

Oczywiście na codzień większość pracy, jeśli idzie o kontrolę wersji naszego kodu, załatwiają za nas aplikacje typu wspomniany wyżej [GitHub Desktop](https://desktop.github.com/), (czy wtyczki do popularnych IDE jak [Atom](https://atom.io/) oraz [PyCharm](https://www.jetbrains.com/pycharm/).

Warto również zapoznać się z ideą pliku `.gitignore` (np. [gitignore dla Python-a](https://github.com/github/gitignore/blob/master/Python.gitignore)).
